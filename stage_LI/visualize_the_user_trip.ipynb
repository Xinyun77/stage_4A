## This part was made for drawing users' trajectory by using ArcGis

```python
# function takes the dataframe of a user as input
# returns the datarame containing lat lon of nodes on the trajectory
# and the dataframe containing lat lon of links to nodes
def dataArcgis(df_user):
    Position = df_user.POSITION.tolist()
    tmp1 = [i.split() for i in Position]
    position_to_latlon =  [[utm.to_latlon(float(x[0]), float(x[1]), 31, 'U')[0], utm.to_latlon(float(x[0]), float(x[1]), 31, 'U')[1]] for x in tmp1]
    lat = [i[0] for i in position_to_latlon]
    lon = [i[1] for i in position_to_latlon]
    X_Start = lat.copy()
    Y_Start = lon.copy()
    X_End = lat.copy()
    Y_End = lon.copy()
    X_Start.pop(-1)
    Y_Start.pop(-1)
    X_End.pop(0)
    Y_End.pop(0)
    df_latlon = pd.DataFrame(list(zip(lat, lon)), columns = ['Latitude', 'Longitude'])
    df_toline = pd.DataFrame(list(zip(X_Start, Y_Start, X_End, Y_End)), columns = ['X_Start', 'Y_Start', 'X_End', 'Y_End'])
    return df_latlon, df_toline
```
